```python
#class to store the matrix for 8-puzzle
class puzzle():
    a=[[0,0,0],
       [0,0,0],
       [0,0,0]]
```

```python
#class to make a move if the empty piece is the center
def center(ini):
    final=[]
    oini=[row[:] for row in ini]
    for val in range(4):
        if(val is 0):
            temp=ini[0][1]
            ini[0][1]=ini[1][1]
            ini[1][1]=temp
        elif(val is 1):
            temp=ini[1][0]
            ini[1][0]=ini[1][1]
            ini[1][1]=temp
        elif(val is 2):
            temp=ini[1][2]
            ini[1][2]=ini[1][1]
            ini[1][1]=temp
        else:
            temp=ini[2][1]
            ini[2][1]=ini[1][1]
            ini[1][1]=temp
        final.append(ini)
        ini=[row[:] for row in oini]
    return final
```

```python
#class to make a move if the empty piece is the corner
def corner(ini,num):
    final=[]
    oini=[row[:] for row in ini]
    for val in range(2):
        if(num is 0 and val is 0):
            temp=ini[0][1]
            ini[0][1]=ini[0][0]
            ini[0][0]=temp
        if(num is 0 and val is 1):
            temp=ini[1][0]
            ini[1][0]=ini[0][0]
            ini[0][0]=temp
        if(num is 1 and val is 0):
            temp=ini[0][1]
            ini[0][1]=ini[0][2]
            ini[0][2]=temp 
        if(num is 1 and val is 1):
            temp=ini[1][2]
            ini[1][2]=ini[0][2]
            ini[0][2]=temp
        if(num is 2 and val is 0):
            temp=ini[1][0]
            ini[1][0]=ini[2][0]
            ini[2][0]=temp
        if(num is 2 and val is 1):
            temp=ini[2][1]
            ini[2][1]=ini[2][0]
            ini[2][0]=temp
        if(num is 3 and val is 0):
            temp=ini[1][2]
            ini[1][2]=ini[2][2]
            ini[2][2]=temp
        if(num is 3 and val is 1):
            temp=ini[2][1]
            ini[2][1]=ini[2][2]
            ini[2][2]=temp
        final.append(ini)
        ini=[row[:] for row in oini]
    return final
```

```python
#class to make a move if the empty piece is the edge
def edge(ini,num):
    final=[]
    oini=[row[:] for row in ini]
    for val in range(3):
        if(num is 0 and val is 0):
            temp=ini[0][0]
            ini[0][0]=ini[0][1]
            ini[0][1]=temp
        if(num is 0 and val is 1):
            temp=ini[0][2]
            ini[0][2]=ini[0][1]
            ini[0][1]=temp
        if(num is 0 and val is 2):
            temp=ini[1][1]
            ini[1][1]=ini[0][1]
            ini[0][1]=temp
        if(num is 1 and val is 0):
            temp=ini[0][0]
            ini[0][0]=ini[1][0]
            ini[1][0]=temp
        if(num is 1 and val is 1):
            temp=ini[1][1]
            ini[1][1]=ini[1][0]
            ini[1][0]=temp
        if(num is 1 and val is 2):
            temp=ini[2][0]
            ini[2][0]=ini[1][0]
            ini[1][0]=temp
        if(num is 2 and val is 0):
            temp=ini[0][2]
            ini[0][2]=ini[1][2]
            ini[1][2]=temp
        if(num is 2 and val is 1):
            temp=ini[1][1]
            ini[1][1]=ini[1][2]
            ini[1][2]=temp
        if(num is 2 and val is 2):
            temp=ini[2][2]
            ini[2][2]=ini[1][2]
            ini[1][2]=temp
        if(num is 3 and val is 0):
            temp=ini[1][1]
            ini[1][1]=ini[2][1]
            ini[2][1]=temp
        if(num is 3 and val is 1):
            temp=ini[2][0]
            ini[2][0]=ini[2][1]
            ini[2][1]=temp
        if(num is 3 and val is 2):
            temp=ini[2][2]
            ini[2][2]=ini[2][1]
            ini[2][1]=temp
        final.append(ini)
        ini=[row[:] for row in oini]
    return final
```

```python
#function to check the alignment of the empty piece
def findmove(ini):
    if(ini[1][1] is 0):
        return 4,0
    elif(ini[0][0] is 0):
        return 2,0
    elif(ini[0][2] is 0):
        return 2,1
    elif(ini[2][0] is 0):
        return 2,2
    elif(ini[2][2] is 0):
        return 2,3
    elif(ini[0][1] is 0):
        return 3,0
    elif(ini[1][0] is 0):
        return 3,1
    elif(ini[1][2] is 0):
        return 3,2
    else:
        return 3,3
```

```python
#function to check repition of any value in already generated tree
def check_rep(ini,f_layer):
    for i in f_layer:
        if(i.a == ini):
            return 1
    return 0
```

```python
#function to add key-value pair in the dictionary
def addlevel(llist):
    if llist[0] not in layer:
        layer[llist[0]]=[]
    for i in range(1,len(llist)):
        layer[llist[0]].append(llist[i])
```

```python
#ini is a static input 
global p
p=0
layer=dict()
obj={}
obj[p]=puzzle()
ini=[[5,1,4],
     [2,0,7],
     [8,3,6]]
oini=[row[:] for row in ini]
obj[p].a=ini
#f_layer holds the key-value pair for one level at a time
f_layer=[]
#BFS holds all the key that have not been generated with their child
BFS=[]
#all_n holds the value all all nodes generated so far
all_n=[]
f_layer.append(obj[p])
BFS.append(obj[p])
all_n.append(obj[p])
p+=1
```

```python
#Creating dictionary using BFS
count=0
x=0
y=0
final=[]
while(BFS):
    if(ini == [[1,2,3],[4,5,6],[7,8,0]]):
        print("\nYOU HAVE A FINAL")
        all_n.append(BFS[0])
        addlevel(f_layer)
        del BFS[0]
        if(len(BFS)==0):
            break
        ini=BFS[0].a
        oini=[row[:] for row in ini]
        f_layer=[]
        f_layer.append(BFS[0])
        continue
    x,y=findmove(oini)
    if(x is 4):
        final=center(oini)
    elif(x is 2):
        final=corner(oini,y)
    else:
        final=edge(oini,y)
    for i in final:
        if((check_rep(i,f_layer)==0) and (check_rep(i,all_n)==0)):
            obj[p]=puzzle() 
            obj[p].a=i
            f_layer.append(obj[p])
            all_n.append(obj[p])
            BFS.append(obj[p])
            p+=1
    addlevel(f_layer)
    del BFS[0]
    if(len(BFS)==0):
        break
    ini=BFS[0].a
    
    oini=[row[:] for row in ini]
    f_layer=[]
    f_layer.append(BFS[0])
    count+=1
    if(count%50==0):
        print(len(BFS),end=",")
```

```python
#Printing dictionary using BFS
for i in layer:
    print(i.a,end=":::")
    for j in layer[i]:
        if(j == layer[i][len(layer[i])-1]):
            print(j.a)
            break
        print(j.a,end=" and ")
    print(end="\n")
```

```python
#finding all possible shortest paths from initial(static input) till final([1,2,3],[4,5,6],[7,8,0]) capacity.
ini=obj[0]
visited=[]
all_p=[]
count=0
flag=0
num_paths=0
while(count==0):
    visited.append(ini)
    all_p.append(ini)
    if(ini.a==[[1,2,3],[4,5,6],[7,8,0]]):
        num_paths+=1
        for i in visited:
            if(i == visited[len(visited)-1]):
                print(i.a,end="\n")
            else:
                print(i.a,end=" and ")
        flag=0
        while(flag==0):
            del(visited[len(visited)-1])
            ini=visited[len(visited)-1]
            for i in layer[ini]:
                if(i not in all_p):
                    ini=i
                    flag=1
                    break
            if(ini==obj[0]):
                flag=1
                count=1
            if(flag==0):
                continue    
    elif(len(layer[ini])==0):
        flag=0
        while(flag==0):
            del(visited[len(visited)-1])
            ini=visited[len(visited)-1]
            for i in layer[ini]:
                if(i not in all_p):
                    ini=i
                    flag=1
                    break
            if(ini==obj[0]):
                flag=1
                count=1
            if(flag==0):
                continue    
    else:
        for i in layer[ini]:
            if(i not in visited):
                ini=i
                break
```
