```python
#class to store the matrix for tic tac toe
class tictactoe():
    a=[[0,0,0],
      [0,0,0],
      [0,0,0]]
```

```python
#function to make a move by any player
def move(ini,val):
    final=[]
    oini=[row[:] for row in ini]
    for i in range(len(ini)):
        for j in range(len(ini[0])):
            if(ini[i][j]== 0):
                ini[i][j]=val
                final.append(ini)
                ini=[row[:] for row in oini]
    return final
```

```python
#function to check repition of any value in already generated tree
def check_rep(ini,f_layer):
    for i in f_layer:
        if(i.a == ini):
            return 1
    return 0
```

```python
#function to add key-value pair in the dictionary
def addlevel(llist):
    if llist[0] not in layer:
        layer[llist[0]]=[]
    for i in range(1,len(llist)):
        layer[llist[0]].append(llist[i])
```

```python
#function to check whether the game has been won or not
def checkResult(ini):
    if((ini[0][0]==ini[0][1]) and (ini[0][0]==ini[0][2]) and (ini[0][0]!=0)):
        return 1
    elif((ini[0][0]==ini[1][0]) and (ini[0][0]==ini[2][0]) and (ini[0][0]!=0)):
        return 1
    elif((ini[0][2]==ini[1][2]) and (ini[0][2]==ini[2][2]) and (ini[0][2]!=0)):
        return 1
    elif((ini[2][0]==ini[2][1]) and (ini[2][0]==ini[2][2]) and (ini[2][0]!=0)):
        return 1
    elif((ini[0][0]==ini[1][1]) and (ini[0][0]==ini[2][2]) and (ini[0][0]!=0)):
        return 1
    elif((ini[0][2]==ini[1][1]) and (ini[0][2]==ini[2][0]) and (ini[0][2]!=0)):
        return 1
    else:
        return 0
```

```python
global p
p=0
obj={}
layer=dict()
obj[p]=tictactoe()
#f_layer holds the key-value pair for one level at a time
f_layer=[]
#BFS holds all the key that have not been generated with their child
BFS=[]
#all_n holds the value all all nodes generated so far
all_n=[]
BFS.append(obj[p])
all_n.append(obj[p])
f_layer.append(obj[p])
p+=1
```

```python
#Creating dictionary using BFS
c=1
count=0
final=[]
ini=[[0,0,0],[0,0,0],[0,0,0]]
oini=[row[:] for row in ini]
while(BFS):
    if(checkResult(ini)==1):
        print("\nYOU HAVE A FINAL")
        all_n.append(BFS[0])
        addlevel(f_layer)
        del BFS[0]
        if(len(BFS)==0):
            break
        ini=BFS[0].a
        oini=[row[:] for row in ini]
        f_layer=[]
        f_layer.append(BFS[0])
        continue
    j=0    
    for i in ini:
        if(0 not in i):
            j+=1
    if(j == 3):
        all_n.append(BFS[0])
        addlevel(f_layer)
        del BFS[0]
        if(len(BFS)==0):
            break
        ini=BFS[0].a
        oini=[row[:] for row in ini]
        f_layer=[]
        f_layer.append(BFS[0])
        continue
    final=move(oini,c)
    for i in final:
        if((check_rep(i,f_layer)==0) and (check_rep(i,all_n)==0)):
            obj[p]=tictactoe() 
            obj[p].a=i
            f_layer.append(obj[p])
            all_n.append(obj[p])
            BFS.append(obj[p])
            p+=1
    addlevel(f_layer)
    del BFS[0]
    if(len(BFS)==0):
        break
    ini=BFS[0].a
    oini=[row[:] for row in ini]
    f_layer=[]
    f_layer.append(BFS[0])
    if(count%50==0):
        print(len(BFS),end=",")
    c1=0
    c2=0
    for i in range(len(ini)):
        for j in range(len(ini[0])):
            if(ini[i][j]==1):
                c1+=1
            elif(ini[i][j]==2):
                c2+=1
    if(c1<=c2):
        c=1
    else:
        c=2
    count+=1
```

```python
#Printing dictionary using BFS
for i in layer:
    print(i.a,end=":::")
    for j in layer[i]:
        if(j == layer[i][len(layer[i])-1]):
            print(j.a)
            break
        print(j.a,end=" and ")
    print(end="\n")
```

```python
#finding all possible shortest paths from initial(static input) till final(winning rules for tic tac toe) capacity.
ini=obj[0]
visited=[]
all_p=[]
count=0
flag=0
num_paths=0
while(count==0):
    visited.append(ini)
    all_p.append(ini)
    if(checkResult(ini.a)==1):
        num_paths+=1
        for i in visited:
            if(i == visited[len(visited)-1]):
                print(i.a,end="\n")
            else:
                print(i.a,end=" and ")
        flag=0
        while(flag==0):
            del(visited[len(visited)-1])
            ini=visited[len(visited)-1]
            for i in layer[ini]:
                if(i not in all_p):
                    ini=i
                    flag=1
                    break
            if(ini==obj[0]):
                flag=1
                count=1
            if(flag==0):
                continue    
    elif(len(layer[ini])==0):
        flag=0
        while(flag==0):
            del(visited[len(visited)-1])
            ini=visited[len(visited)-1]
            for i in layer[ini]:
                if(i not in all_p):
                    ini=i
                    flag=1
                    break
            if(ini==obj[0]):
                flag=1
                count=1
            if(flag==0):
                continue    
    else:
        for i in layer[ini]:
            if(i not in visited):
                ini=i
                break
```
